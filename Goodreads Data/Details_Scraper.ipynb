```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url_to_fetch = "https://www.goodreads.com/book/show/61851.Alcatraz_Versus_the_Evil_Librarians" # Example URL
response = requests.get(url_to_fetch)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
book_data = [] # Run this cell before running the scraping loop
```

```python
# Define the list of audience/age categories
audience_categories = ["Picture Book", "Early Reader", "Middle Grade", "Young Adult", "New Adult", "Adult"]
```

```python
for url in urls: # Use the urls list defined in cell b081e051
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract the book title
    title_h1 = soup.find('h1', {'data-testid': 'bookTitle'})
    title = title_h1.get_text(strip=True) if title_h1 else "Title not found"

    # Extract the book description
    description_div = soup.find('div', {'data-testid': 'description'})
    description = description_div.get_text(strip=True) if description_div else "Description not found"

    # Extract the published date
    published_date_p = soup.find('p', {'data-testid': 'publicationInfo'})
    published_date = published_date_p.get_text(strip=True) if published_date_p else "Published date not found"

    # Extract genre buttons and check against audience categories
    genres_in_audience = []
    non_audience_genres = []
    genre_buttons = soup.find_all('span', class_='BookPageMetadataSection__genreButton')
    for genre_button in genre_buttons:
        genre_span = genre_button.find('span', class_='Button__labelItem')
        if genre_span:
            genre = genre_span.get_text(strip=True)
            if genre in audience_categories:
                genres_in_audience.append(genre)
            else:
                non_audience_genres.append(genre)


    # Extract currently reading count
    currently_reading_div = soup.find('div', {'data-testid': 'currentlyReadingSignal'})
    currently_reading = currently_reading_div.get_text(strip=True).split()[0] if currently_reading_div else "Not found"

    # Extract want to read count
    to_read_div = soup.find('div', {'data-testid': 'toReadSignal'})
    to_read = to_read_div.get_text(strip=True).split()[0] if to_read_div else "Not found"

    # Extract average rating
    average_rating_div = soup.find('div', class_='RatingStatistics__rating')
    average_rating = average_rating_div.get_text(strip=True) if average_rating_div else "Not found"

    # Extract total ratings count
    ratings_count_span = soup.find('span', {'data-testid': 'ratingsCount'})
    ratings_count = ratings_count_span.get_text(strip=True).split()[0] if ratings_count_span else "Not found"

    # Extract total review count
    reviews_count_span = soup.find('span', {'data-testid': 'reviewsCount'})
    reviews_count = reviews_count_span.get_text(strip=True).split()[0] if reviews_count_span else "Not found"

    # Extract author name
    author_span = soup.find('span', {'data-testid': 'name'})
    author = author_span.get_text(strip=True) if author_span else "Not found"

    book_data.append({
        "URL": url,
        "Title": title,
        "Description": description,
        "Published Date": published_date,
        "Audience Genre": ", ".join(genres_in_audience) if genres_in_audience else "Not specified",
        "Non-Audience Genres": ", ".join(non_audience_genres) if non_audience_genres else "Not specified",
        "Currently Reading": currently_reading,
        "Want to Read": to_read,
        "Average Rating": average_rating,
        "Total Ratings": ratings_count,
        "Total Reviews": reviews_count,
        "Author": author

    })
```

```python
book_df = pd.DataFrame(book_data)
book_df['Published Date'] = book_df['Published Date'].str.replace('First published ', '', regex=False)
book_df['Currently Reading'] = book_df['Currently Reading'].str.replace(',', '', regex=False).astype(int, errors='ignore')
book_df['Want to Read'] = book_df['Want to Read'].str.replace(',', '', regex=False).astype(int, errors='ignore')
book_df['Total Ratings'] = book_df['Total Ratings'].str.replace('ratings', '', regex=False).str.replace(',', '', regex=False).astype(int, errors='ignore')
book_df['Total Reviews'] = book_df['Total Reviews'].str.replace('reviews', '', regex=False).str.replace(',', '', regex=False).astype(int, errors='ignore')
book_df['Average Rating'] = book_df['Average Rating'].astype(float, errors='ignore')


# Reorder columns to include the new 'Non-Audience Genres' column after 'Total Reviews'
book_df = book_df[['Author', 'Title', 'Published Date', 'Audience Genre', 'Description', 'Average Rating', 'Total Ratings', 'Total Reviews', 'Non-Audience Genres', 'Currently Reading', 'Want to Read', 'URL']]

display(book_df)
```
