```python
import pandas as pd
import glob
import os

# Replace with the path to your directory containing the CSV files
csv_directory = '/content/drive/MyDrive/School/MSBA Program/ANLY 6900 - Capstone Project/Goodreads_Scraped_Data/Reviews'
csv_files = glob.glob(f'{csv_directory}/*.csv')

# Create an empty list to store individual DataFrames
dfs = []

# Read each CSV file into a DataFrame, add book title, and append it to the list
for csv_file in csv_files:
    try:
        df = pd.read_csv(csv_file)
        # Extract book title from filename and remove '_reviews' and replace '_' with ' '
        book_title = os.path.splitext(os.path.basename(csv_file))[0].replace('_reviews', '').replace('_', ' ')
        df['book_title'] = book_title  # Add book_title column
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {csv_file}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    display(combined_df.head())
else:
    print("No CSV files found or processed.")
```

There area lot of word_counts that weren't calculated when the original code was run for the web scraper. The code below should get the word_count for the reviews

```python
import numpy as np

# Function to calculate word count
def count_words(text):
    if isinstance(text, str):
        return len(text.split())
    else:
        return np.nan # Return NaN for non-string values

# Apply the function to the 'review_text' column and fill NaN values in 'word_count'
combined_df['word_count'] = combined_df.apply(
    lambda row: count_words(row['review_text']) if pd.isna(row['word_count']) else row['word_count'],
    axis=1
)

# Display the updated DataFrame with filled word counts
display(combined_df.head())
```

```python
# Replace 'output.csv' with your desired output file name and path
output_file = '/content/drive/MyDrive/School/MSBA Program/ANLY 6900 - Capstone Project/Goodreads_Scraped_Data/Combined_Reviews.csv'

# Save the combined DataFrame to a new CSV file
combined_df.to_csv(output_file, index=False)

print(f"Combined data saved to {output_file}")
```
